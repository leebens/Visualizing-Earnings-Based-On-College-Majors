# College Graduate Job Outcome Analysis

In this project I'll be working with a dataset on the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by American Community Survey, which conducts surveys and aggregates the data. FiveThirtyEight cleaned the dataset and released it on their Github repo. I will be experimenting with different visualizations to answer the following questions:

- Do students in more popular majors make more money?
- How many majors are predominantly male? Predominantly female?
- Which category of majors have the most students?

```python
import pandas as pd
import matplotlib
from IPython.display import display
%matplotlib inline 
```

```python
recent_grads = pd.read_csv('recent-grads.csv')

display(recent_grads.iloc[0]) #return the first row

display(recent_grads.head())
display(recent_grads.tail())
display(recent_grads.describe())
```

```python
raw_data_count = recent_grads.shape[0] #number of rows of data
print(raw_data_count)

recent_grads = recent_grads.dropna()
print(recent_grads.shape[0])

print('\n')
print('We can see that only one row was dropped that contained missing values. Matplotlib expects that columns of values we pass in have matching lengths and missing values will cause matplotlib to throw errors.')
```

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Sample_size vs. Median', figsize=(5,10))
```

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Sample_size vs. Unemplyment_rate', figsize=(5,10))
```

From the scatter plot above we do not necessarily see an indication that more popular majors make more money. Potentially there could be a bell curve, we will need to investigate more. 

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Full_time vs. Median', figsize=(5,10))
```

From the above graph we can see that it is possible there is a potentially inverse relationship between full-time employees and median salary.

```python
recent_grads.plot(x='Unemployment_rate', y='ShareWomen', kind='scatter', title='ShareWomen vs. Unemplyment_rate', figsize=(5,10))
```

```python
recent_grads.plot(x='Men', y='Median', kind='scatter', title='Men vs. Median', figsize=(5,10))
```

```python
recent_grads.plot(x='Women', y='Median', kind='scatter', title='Women vs. Median', figsize=(5,10))
```
